```python
#Marksheet
flag = True
#Using while loop to get multiple users input
while flag:
    dictn = {}
#While loop for multiple subjects input
    while True:
        sub1 = input("Please enter the name of the subject: ")
        marks1 = int(input('Please enter the marks of the subject: '))
        dictn[sub1] = marks1
        b = input("\nPress enter to continue or type 'exit' to quit: ")
        if b == 'exit':
            break
        else:
            continue
#Displaying subject and marks at the output     
    print("\n")
    for k,v in dictn.items():
        if len(k) < 8:
            print(k.title() + "\t\t\t" + str(v))
        elif len(k) >= 8 and len(k) < 16:
            print(k.title() + "\t\t" + str(v))
        else:
            print(k.title() + "\t" + str(v))
    dictn
    total1 = sum(dictn.values())
#Calculating total marks
    print("\nThe sum of marks are: " + str(total1))
    sc = len(dictn)
    per = int(total1 / sc)
#Determining percentage
    print("\nYour obtained percentage is: " + str(per) + "%.\n")
#Checking grade    
    if per <= 100 and per >= 0:
        if per >= 90 and per <= 100:
            print("Great! You have got A+ grade.")
        elif per >= 80 and per < 90:
            print("You have got A grade.")
        elif per >= 70 and per < 80:
            print("You have got B+ grade.")
        elif per >= 60 and per < 70:
            print("You have got B grade.")
        elif per >= 50 and per < 60:
            print("You have got C grade.")
        elif per < 50 and per >= 0:
            print("Alas! You have got F grade.")
        elif per > 100:
            print("Invalid grade.")
        else:
            print("Invalid number.")
    else:
        print("Please enter the marks that is less than the total marks and greater than 0.")
    a = input("\nPress enter to continue or type 'exit' to quit: ")
    if a == 'exit':
        flag = False
    else:
        continue
```
